## Transformaciones

### Transformaciones Multifiltro

```python
# 1. Eliminar columnas innecesarias
columnas_a_eliminar = [
    'comment', 'VTEX ID', 'tracking_number',
    'ship_to_warehouse_name', 'ship_to_warehouse_id',
    'doc_type', 'doc_number', 'email', 'OMS ID',
    'Precio Total', 'store_id', 'store_name',
    'zip_code', 'title', 'unit_price', 'discount'
]

Multrifiltro  = Multifiltro.drop(columns=columnas_a_eliminar, errors='ignore')

# 2. Crear columna de estado general del pedido
def clasificar_estado(row):
    if row['status'] == 'delivered':
        return 'Entregado'
    elif pd.notnull(row['cancelled_at']):
        return 'Cancelado'
    else:
        return 'Pendiente'

Multifiltro['Estado General'] = Multifiltro.apply(clasificar_estado, axis=1)

# 3. Convertir fechas a datetime
Multifiltro['created_at'] = pd.to_datetime(Multifiltro['created_at'], errors='coerce')
Multifiltro['delivered_at'] = pd.to_datetime(Multifiltro['delivered_at'], errors='coerce')
Multifiltro['cancelled_at'] = pd.to_datetime(Multifiltro['cancelled_at'], errors='coerce')

```

```python
## Eliminar Columnas
columnas_a_eliminar = [
    'OMS ID', 'Precio Total', 'status', 'store_id',
    'store_name', 'zip_code', 'title', 'unit_price',
    'discount','comment', 'tracking_number','ship_to_warehouse_name' ,
    'ship_to_warehouse_id', 'doc_type', 'doc_number', 'email','VTEX ID'
]
Multifiltro = Multifiltro.drop(columns=columnas_a_eliminar)

```

```python
## Renombrar Columna Articulo
Multifiltro = Multifiltro.rename(columns={'Referencia externa': 'Articulo'})

## Combinar Nombre Completo
Multifiltro['Nombre completo'] = Multifiltro['name'].astype(str) + ' ' + Multifiltro['last_name'].astype(str)
Multifiltro = Multifiltro.drop(columns=['name', 'last_name'])
```

## EDA

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el dataset
df = pd.read_csv("Archivos Transformados/Informe_Pendientes_Online.csv")

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 4)  # Más compacto por fila

# Función para agregar etiquetas encima de las barras
def add_value_labels(ax, spacing=5):
    """Agrega etiquetas numéricas a un gráfico de barras"""
    for p in ax.patches:
        if p.get_height() > 0:
            ax.annotate(f'{int(p.get_height())}',
                        (p.get_x() + p.get_width() / 2., p.get_height()),
                        ha='center', va='bottom',
                        fontsize=8, color='black',
                        xytext=(0, spacing),
                        textcoords='offset points')

# =================================
# 1. VALORES NULOS
# =================================
plt.figure(figsize=(12, 4))
sns.heatmap(df.isnull(), cbar=False, cmap="Blues")
plt.title("Valores nulos en el DataFrame")
plt.tight_layout()
plt.show()

# =================================
# 2. DISTRIBUCIÓN DE VARIABLES NUMÉRICAS
# =================================
numeric_cols = df.select_dtypes(include=["float64", "int64"]).columns

for i in range(0, len(numeric_cols), 2):
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    for j in range(2):
        if i + j < len(numeric_cols):
            sns.histplot(df[numeric_cols[i + j]], kde=True, ax=axes[j], bins=30)
            axes[j].set_title(f"Distribución de {numeric_cols[i + j]}", fontsize=10)
    plt.tight_layout()
    plt.show()

# =================================
# 3. VARIABLES CATEGÓRICAS CON VALORES
# =================================
cat_cols = df.select_dtypes(include=["object"]).nunique()
cat_cols = cat_cols[cat_cols < 30].index.tolist()

for i in range(0, len(cat_cols), 2):
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    for j in range(2):
        if i + j < len(cat_cols):
            col = cat_cols[i + j]
            order = df[col].value_counts().index
            ax = sns.countplot(data=df, x=col, order=order, ax=axes[j])
            axes[j].set_title(f"{col}", fontsize=10)
            axes[j].tick_params(axis='x', rotation=45)
            add_value_labels(ax)
    plt.tight_layout()
    plt.show()

# =================================
# 4. MATRIZ DE CORRELACIÓN
# =================================
plt.figure(figsize=(10, 6))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriz de correlación entre variables numéricas")
plt.tight_layout()
plt.show()

# =================================
# 5. MULTIVARIABLES EJEMPLO CON VALORES
# =================================

# Estado General vs Estado Devolución
plt.figure(figsize=(12, 4))
ax = sns.countplot(data=df, x="Estado General", hue="Estado Devolución",
                   order=df["Estado General"].value_counts().index)
plt.title("Estado General vs Estado Devolución")
plt.xticks(rotation=45)
add_value_labels(ax)
plt.tight_layout()
plt.show()

# Acción Programada vs Estado Devolución
plt.figure(figsize=(12, 4))
ax = sns.countplot(data=df, x="Acción Programada", hue="Estado Devolución",
                   order=df["Acción Programada"].value_counts().index)
plt.title("Acción Programada vs Estado Devolución")
plt.xticks(rotation=45)
add_value_labels(ax)
plt.tight_layout()
plt.show()

# Quantity vs Estado Devolución (Boxplot no lleva etiquetas)
if "quantity" in df.columns:
    plt.figure(figsize=(12, 4))
    sns.boxplot(data=df, x="Estado Devolución", y="quantity")
    plt.title("Cantidad por Estado de Devolución")
    plt.tight_layout()
    plt.show()
```

```python
# Limpiar columnas innecesarias del archivo de recepciones
recepciones_limpio = Recepciones[[
    'Fecha creación',
    'Código',
    'Clase de entrada',
    'Tipo de recepción',
    'Estado'
]]

def extraer_orden_entrada(valor):
    if isinstance(valor, str) and valor.startswith("RK_"):
        return "_".join(valor.split("_")[:4])  # ejemplo: RK_RMC_3333_00000010
    else:
        return str(valor)[:8]  # primeros 8 caracteres para valores numéricos

recepciones_limpio['Orden de entrada'] = recepciones_limpio['Código'].apply(extraer_orden_entrada)
```

### Stock Recibido




```python
# === 1) ELIMINAR COLUMNA "Proveedor" ===
stock_recibido_limpio = Stock_Recibido.drop(columns=['Proveedor'], errors='ignore').copy()

# === 2) FORMATEAR "Fecha recepción" + SEPARAR EN "Fecha" y "Hora" ===
stock_recibido_limpio['Fecha recepción'] = pd.to_datetime(
    stock_recibido_limpio['Fecha recepción'], errors='coerce', dayfirst=True
)
stock_recibido_limpio['Fecha'] = stock_recibido_limpio['Fecha recepción'].dt.date
stock_recibido_limpio['Hora'] = stock_recibido_limpio['Fecha recepción'].dt.time
stock_recibido_limpio.drop(columns=['Fecha recepción'], inplace=True)

# === 3) EXTRAER "Orden de entrada" DESDE 'Recepción' ===
def extraer_orden_entrada(valor):
    if pd.isna(valor):
        return pd.NA
    s = str(valor)
    if s.startswith('RK_'):
        parts = s.split('_')
        if len(parts) >= 4:
            prefijo = '_'.join(parts[:3])
            segmento = parts[3][:8]
            return f"{prefijo}_{segmento}"
        return s[:8]
    return s[:8]

stock_recibido_limpio['Orden de entrada'] = stock_recibido_limpio['Recepción'].apply(extraer_orden_entrada)

# === 4) GUARDAR ARCHIVO TRANSFORMADO ===
stock_recibido_limpio.to_csv("Archivos Transformados/Stock_Recibido_Transformado.csv", index=False, encoding="utf-8-sig")

# (Opcional) Vista rápida
print(stock_recibido_limpio.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el consolidado limpio con nombre correcto
df = pd.read_csv("Archivos Transformados/Informe_Entradas_Mecalux.csv", parse_dates=["Creacion", "Fecha recepción"])

# Preprocesamiento de columnas necesarias
df['Fecha'] = df['Creacion'].dt.date
df['Hora'] = pd.to_datetime(df['Hora'], errors='coerce').dt.hour
df['Mes'] = df['Creacion'].dt.strftime('%B')  # Nombre del mes
df['Mes'] = pd.Categorical(df['Mes'], categories=[
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
], ordered=True)

# Dividir columna 'Artículo' para extraer Modelo
if 'Artículo' in df.columns:
    df[['Modelo', 'ColorTalle']] = df['Artículo'].astype(str).str.split('%', expand=True)
    df['ColorTalle'] = df['ColorTalle'].fillna('')
    df['Color'] = df['ColorTalle'].str[:2]
    df['Talle'] = df['ColorTalle'].str[2:]
    df.drop(columns='ColorTalle', inplace=True)

# Agrupamientos para los gráficos actualizados
entradas_por_mes = df['Mes'].value_counts().sort_index()
top_modelos = df['Modelo'].value_counts().head(10)
cantidad_por_modelo = df.groupby('Modelo')['Cantidad'].sum().sort_values(ascending=False).head(10)
entradas_por_hora = df['Hora'].value_counts().sort_index()
estado_stock = df['Estado de stock'].value_counts()

# Crear subplots
fig, axs = plt.subplots(3, 2, figsize=(16, 12))
fig.suptitle("Visualización Exploratoria - Informe Entradas Mecalux", fontsize=16, y=1.02)

# 1. Entradas por Mes
axs[0, 0].bar(entradas_por_mes.index, entradas_por_mes.values, color='steelblue')
axs[0, 0].set_title("Entradas por Mes")
axs[0, 0].set_xlabel("Mes")
axs[0, 0].set_ylabel("Cantidad")
axs[0, 0].tick_params(axis='x', rotation=45)

# 2. Top modelos ingresados (frecuencia)
axs[0, 1].barh(top_modelos.index[::-1], top_modelos.values[::-1], color='skyblue')
axs[0, 1].set_title("Top 10 Modelos Más Ingresados")
axs[0, 1].set_xlabel("Cantidad")
axs[0, 1].set_ylabel("Modelo")

# 3. Cantidad total ingresada por modelo
axs[1, 0].bar(cantidad_por_modelo.index, cantidad_por_modelo.values, color='orange')
axs[1, 0].set_title("Top 10 Modelos por Cantidad Total")
axs[1, 0].set_ylabel("Cantidad Total")
axs[1, 0].set_xlabel("Modelo")
axs[1, 0].tick_params(axis='x', rotation=45)

# 4. Entradas por hora
axs[1, 1].bar(entradas_por_hora.index, entradas_por_hora.values, color='green')
axs[1, 1].set_title("Distribución de Entradas por Hora")
axs[1, 1].set_xlabel("Hora del Día")
axs[1, 1].set_ylabel("Cantidad")
axs[1, 1].set_xticks(range(0, 24))

# 5. Estado de stock (pie chart)
axs[2, 0].pie(estado_stock, labels=estado_stock.index, autopct='%1.1f%%', startangle=90)
axs[2, 0].set_title("Distribución por Estado de Stock")

# Ocultar subplot vacío (último)
axs[2, 1].axis('off')

plt.tight_layout()
plt.subplots_adjust(top=0.92)
plt.show()


plt.tight_layout()
plt.subplots_adjust(top=0.92)
plt.show()
```
